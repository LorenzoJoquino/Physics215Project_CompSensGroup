```python
import numpy as np
from numpy import pi
from scipy.fft import fft, ifft, fftfreq
import matplotlib.pyplot as plt
```

```python
sampling_rate = 2000
time = np.arange(0,1,1/sampling_rate)

sr_random = 128
time_random = np.random.rand(sr_random)
time_random = np.sort(time_random)
```

```python
def signal(time):
    return np.sin(73*2*pi*time) + np.sin(531*2*pi*time)

signal = np.sin(73*2*pi*time) + np.sin(531*2*pi*time)

signal_random = [1]
```

```python
signal_fft = fft(signal)
N = len(signal)
signal_fft_mag = np.abs(signal_fft)/N
freq = fftfreq(N, d = 1/sampling_rate)

half_N = N // 2
X_mag = 2 * signal_fft_mag[:half_N]
freq = freq[:half_N]
```

```python
plt.plot(time,signal)
plt.show()
plt.plot(freq,X_mag.real)
plt.show()
```

```python
import numpy as np
from numpy import pi
from scipy.fft import fft, ifft, fftfreq
import matplotlib.pyplot as plt
```

```python
def signal(time):
    return np.sin(12*2*pi*time) + np.sin(120*2*pi*time)

def fourier(signal, sampling_rate):
    signal_fft = fft(signal)
    N = len(signal)
    signal_fft_mag = np.abs(signal_fft)/N
    freq = fftfreq(N, d = 1/sampling_rate)

    half_N = N // 2
    X_mag = 2 * signal_fft_mag[:half_N]
    freq = freq[:half_N]

    indices = np.argsort(X_mag)[-2:]
    return freq, X_mag.real, freq[indices]
```

```python
sampling_rate = 121
time = np.arange(0,1,1/sampling_rate)

sampling_rate_rand = 100
time_rand = np.random.rand(sampling_rate_rand)
time_rand = np.sort(time_rand)

sig_norm = signal(time)
sig_rand = signal(time_rand)
freq_norm, sigfft_norm, max_freq = fourier(sig_norm, sampling_rate)
freq_rand, sigfft_rand, max_freq_rand = fourier(sig_rand, sampling_rate_rand)
plt.plot(time_rand, sig_rand)
plt.show()
plt.plot(time,sig_norm)
plt.show()
```
